#### 2.2 Fun Experiment on two qubit rotation

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.u3(np.pi/4,np.pi/4,0,1)
qc.barrier()
for i in range(8):
    qc.rz(np.pi/4, 0) 
    qc.rz(np.pi/4, 1)
    
qc.barrier()
qc.h(0)
qc.u3(-np.pi/4,-np.pi/4,0,1)
style = {'backgroundcolor': 'lavender'}
qc.draw(output='mpl', style = style)
```

### Special $R_{\phi}$ Gates: 

$R_{\phi}$ gate is reprisented by $S$ gate when $\phi = \pi/2$. It is also called $\sqrt{Z}$ gate.

$S = \begin{bmatrix} 1 & 0 \\ 0 & e^{\frac{i\pi}{2}} \end{bmatrix}, \quad  S^\dagger = \begin{bmatrix} 1 & 0 \\ 0 & e^{-\frac{i\pi}{2}} \end{bmatrix}$

<img src="quantumgates.png" width="300" hight="300" />

$R_{\phi}$ gate is reprisented by $T$ gate when $\phi = \pi/4$.

$T = \begin{bmatrix} 1 & 0 \\ 0 & e^{\frac{i\pi}{4}} \end{bmatrix}, \quad  T^\dagger = \begin{bmatrix} 1 & 0 \\ 0 & e^{-\frac{i\pi}{4}} \end{bmatrix}$
